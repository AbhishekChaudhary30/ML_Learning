```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
df.head(3)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
# missing values in particularly row
df[df.isnull().any(axis = 1)]
```

```python
df.loc[:122,'Region'] = 0
df.loc[122:,'Region'] = 1
df.head()
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Removing the null values
df = df.dropna().reset_index(drop = True)
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# remove the 122nd row
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# fix white spaces in columns name
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
# convert datatype object into int.
df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype(int)
df.info()
```

```python
# convert dtypes object into float use loop(for).

objects = [features for features in df.columns if df[features].dtypes == 'O']

for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)

df.info()
```

```python
# Let save the cleaned dataset file into a new file, so that we can directly apply ML algos.
df.to_csv('Algerian_forest_fires_dataset_.csv', index = False)
```
