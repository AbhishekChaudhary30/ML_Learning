```python
# We are woking on project --> Bengaluru_House_Data
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.preprocessing import LabelEncoder
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("car data.csv")
df.head(3)
```

```python
# Look data --> Target is selling price
```

```python
df.isnull().sum()
```

```python
categorical = df.select_dtypes(include = "O")
categorical
```

```python
le = LabelEncoder()
```

```python
for i in categorical:
    df[i] = le.fit_transform(df[i])
```

```python
df.head(3)
```

```python
from scipy.stats import zscore
z = np.abs(zscore(df))
df1 = df[(z<3).all(axis = 1)]
```

```python
df.shape
```

```python
df1.shape
```

```python
df = df1.copy()
```

```python
df.shape
```

```python
x = df.drop("Selling_Price", axis = 1)
y = df['Selling_Price']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_sc = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_sc, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
cal = pd.DataFrame(np.c_[y_test, y_pred], columns=['Actual Selling Price', "Predict Selling Price"])
```
