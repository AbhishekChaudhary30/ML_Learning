```python
df = pd.read_csv("Travel.csv")
df.head()
```

```python
df.isnull().sum()  # This line of code is used for checking null values in the data.
```

```python
df.isna().sum() # This line of code is used for checking null values in the data.
```

```python
# For Find out Null Values in the data ----> 1. df.isnull().sum()     2. df.isna().sum()
```

```python
# check all categories 
df["Gender"].value_counts()  # here is a problem that is in Fe Male part of Gender. So next we replace it in the Female
```

```python
df["MaritalStatus"].value_counts() # here is a problem that is 
```

```python
df["Gender"] = df["Gender"].replace("Fe Male", "Female")
df["MaritalStatus"] = df["MaritalStatus"].replace("Single", "Unmarried")
```

```python
df["Gender"].value_counts()
```

```python
df["MaritalStatus"].value_counts()
```

```python
df.info()
```

```python
# filling data in missing values ---- for missing values check -> df.isnull().sum()
# Age
df.Age.fillna(df.Age.median(), inplace = True)

# TypeofContact
df.TypeofContact.fillna(df.TypeofContact.mode()[0], inplace = True)

# DurationOfPitch
df.DurationOfPitch.fillna(df.DurationOfPitch.median(), inplace = True)

# NumberOfFollowups
df.NumberOfFollowups.fillna(df.NumberOfFollowups.mode()[0], inplace=True)

# PreferredPropertyStar
df.PreferredPropertyStar.fillna(df.PreferredPropertyStar.mode()[0], inplace=True)

# NumberOfTrips
df.NumberOfTrips.fillna(df.NumberOfTrips.median(), inplace=True)

# NumberOfChildrenVisiting
df.NumberOfChildrenVisiting.fillna(df.NumberOfChildrenVisiting.mode()[0], inplace=True)

#MonthlyIncome
df.MonthlyIncome.fillna(df.MonthlyIncome.median(), inplace=True)
```

```python
df.head()
df.isnull().sum()
```

```python
df.drop("CustomerID", inplace=True, axis=1) # Q1. Why did we drop "CustomerID" ?
```

```python
df.head()
```

```python
# Get all the numeric features            
num_features = [feature for feature in df.columns if df[feature].dtype !="O"]
print("Num of Numerical Features :", len(num_features))

# Q What is feature?
```

```python
# Get all the categorical features
cat_features = [feature for feature in df.columns if df[feature].dtype =="O"]
print("Num of Categorical Features :", len(cat_features))
```

```python
# Get all the Discrete features ---> It is a subset of Numeric Features
discrete_features = [feature for feature in num_features if len (df[feature].unique())<=25]
print("Num of Discrete Features :", len(discrete_features))

# Q How this line of code 
```

```python
# Continuous features  --->  It is a subset of Numeric Features
continuous_features = [feature for feature in num_features if feature not in discrete_features]
print("Num of Continuous Features :", len(continuous_features))
```

```python
df = pd.read_csv("placement.csv")
df.head()
```

```python
# Topic is Detect Outlier and Removal Outlier
```

```python
df.isnull().sum()
```

```python
# Check Outlier ---> make distribution plot for checking sequence. It means sequence is present, so we can say that Outlier is also present in data.
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.distplot(df['cgpa'])

plt.subplot(1,2,2)
sns.distplot(df["placement_exam_marks"])

plt.show()
```

```python
# Checking for Outliers for cgpa column  -- making boxplot graph 
sns.boxplot(df["cgpa"])
```

```python
# Find Lower Quartile (Q1)
q1_cgpa = df['cgpa'].quantile(0.25)

# Find Upper Quartile (Q3) 
q3_cgpa = df['cgpa'].quantile(0.75)

# Printing Results
print("Q1", q1_cgpa)
print("Q3", q3_cgpa)
print("IQR :", q3_cgpa - q1_cgpa)
```

```python
# Lets find no. of outliers
# for that we have to find the upper and lower outlier limit
outlier_lower_limit = q1_cgpa - 1.5 * (q3_cgpa - q1_cgpa)
outlier_upper_limit = q3_cgpa + 1.5 * (q3_cgpa - q1_cgpa)
print("outlier upper limit :", outlier_upper_limit)
print("outlier lower limit :", outlier_lower_limit)
```

```python
df[df["cgpa"]>outlier_upper_limit] # this line tell us upper outliers
```

```python
df[df["cgpa"]<outlier_lower_limit] # this line tell us lower outliers
```

```python
data_df = df[(df.cgpa>outlier_lower_limit) & (df.cgpa<outlier_upper_limit)] # This line of code tell us, legal data between lower and upper 
```

```python
data_df.head()
```

```python
sns.boxplot(data_df['cgpa'])
```
