```python
# There are Column names where present missing values.
# Age, TypeofContact, DurationOfPitch, NumberOfFollowups, PreferredPropertyStar, NumberOfTrips, NumberOfChildrenVisiting, MonthlyIncome.
```

```python
# Filling missing data

# Age
df.Age.fillna(df.Age.median(), inplace = True)

# TypeofContact
df.TypeofContact.fillna(df.TypeofContact.mode()[0], inplace =  True) 

# DurationOfPitch
df.DurationOfPitch.fillna(df.DurationOfPitch.median(), inplace=True)

# NumberOfFollowups
df.NumberOfFollowups.fillna(df.NumberOfFollowups.mode()[0], inplace = True)

# PreferredPropertyStar
df.PreferredPropertyStar.fillna(df.PreferredPropertyStar.mode()[0], inplace = True)

# NumberOfTrips
df.NumberOfTrips.fillna(df.NumberOfTrips.median(), inplace = True)

# NumberOfChildrenVisiting
df.NumberOfChildrenVisiting.fillna(df.NumberOfChildrenVisiting.mode()[0], inplace = True)

# MonthlyIncome
df.MonthlyIncome.fillna(df.MonthlyIncome.median(), inplace=True)
```

```python
df.isna().sum()
```

```python
df.drop("CustomerID", inplace=True, axis=1)
```

```python
df.head(3)
```

```python
# Now here we will see how numerical values and categorical values are separated from the data?
```

```python
# Get all the numeric features -----------------------> Numeric Values
num_features = [feature for feature in df.columns if df[feature].dtype != "O"]
print("Num of Numerical Features :", len(num_features))
```

```python
# Get all the categoric features -----------------------> Categorical Values
cat_features = [feature for feature in df.columns if df[feature].dtype == "O"]
print("Num of Categorical Features", len(cat_features))
```

```python
# Get all the Discrete features --> It is a subset of Numeric Features
discrete_features = [feature for feature in num_features if len(df[feature].unique()) <= 25]
print("Num of Discrete Features :", len(discrete_features))
```

```python
# Continuous features --> Subset of Numeric Features.
continuous_features = [feature for feature in num_features if feature not in discrete_features]
print("Num of Continuous Features: ", len(continuous_features))
```

```python
df = pd.read_csv("placement.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
# If there is no missing values present in data then we move ahead.
```

```python
# making distribution plot
# To check sequence of Outliers.
plt.figure(figsize = (16,5))

plt.subplot(1,2,1)
sns.distplot(df['cgpa'])

plt.subplot(1,2,2)
sns.distplot(df['placement_exam_marks'])
```

```python
# Now making boxplot
sns.boxplot(df["cgpa"])
```

```python
sns.boxplot(df['placement_exam_marks'])
```

```python
# Step 1
# Find Quaritle  for "cgpa" --> (Lower, Upper and IQR)
# Find Lower Quartile (Q1)
q1_cgpa = df['cgpa'].quantile(0.25)

# Find Upper Quaritle (Q3)
q3_cgpa = df['cgpa'].quantile(0.75)

# Find "IQR"
IQR_cgpa = q3_cgpa - q1_cgpa


# printing
print("Q1 is: ", q1_cgpa)
print("Q3 is: ", q3_cgpa)
print("IQR is: ", IQR_cgpa)
```

```python
# Find Quaritle for "placement_exam_marks" --> (Lower, Upper and IQR)
# Lower Quaritle (Q1)
q1_placement_exam_marks = df['placement_exam_marks'].quantile(0.25)

# Upper Quaritle (Q3)
q3_placement_exam_marks = df['placement_exam_marks'].quantile(0.75)

# Find "IQR"
IQR_placement_exam_marks = q3_placement_exam_marks - q1_placement_exam_marks



# printing
print("Q1 is: ", q1_placement_exam_marks)
print("Q3 is: ", q3_placement_exam_marks)
print("IQR is: ", IQR_placement_exam_marks)
```

```python
# Step 2
# Now finding Outliers (no. of outliers present) for "cgpa" --> Upper Limit Outliers and Lower Limit Outliers.
outlier_lower_limit_cgpa = q1_cgpa - 1.5 * (IQR_cgpa)
outlier_upper_limit_cgpa = q3_cgpa + 1.5 * (IQR_cgpa)
print("outlier lower limit: ", outlier_lower_limit_cgpa)
print("outlier upper limit: ", outlier_upper_limit_cgpa)
```

```python
# Now finding Outliers for "placement_exam_marks" --> Upper Limit Outliers and Lower Limit Outliers.
outlier_lower_limit_placement_exam_marks = q1_placement_exam_marks - 1.5 * IQR_placement_exam_marks
outlier_upper_limit_placement_exam_marks = q3_placement_exam_marks + 1.5 * IQR_placement_exam_marks
print("outlier lower limit: ", outlier_lower_limit_placement_exam_marks)
print("outlier upper limit: ", outlier_upper_limit_placement_exam_marks)
```

```python
df[df['cgpa'] > outlier_upper_limit_cgpa]
```

```python
df[df['cgpa'] < outlier_lower_limit_cgpa]
```

```python
df[df['placement_exam_marks'] > outlier_upper_limit_placement_exam_marks]
```

```python
df[df['placement_exam_marks'] < outlier_lower_limit_placement_exam_marks]
```

```python
data_df1 = df[(df.cgpa > outlier_lower_limit_cgpa) & (df.cgpa < outlier_upper_limit_cgpa)]
```

```python
data_df2 = df[(df.placement_exam_marks > outlier_lower_limit_placement_exam_marks) & (df.placement_exam_marks < outlier_upper_limit_placement_exam_marks)]
```

```python
data_df1.head(3)
```

```python
data_df2.head(3)
```

```python
sns.boxplot(data_df1['cgpa'])
```

```python
sns.boxplot(data_df2['placement_exam_marks'])
```

```python
# Work is finished ---> Detect Outliers and Remove Outliers using Quaritle method.
##################################################################################################
```
