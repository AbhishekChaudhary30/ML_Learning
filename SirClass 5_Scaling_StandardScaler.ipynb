```python
# Scaling --> to scale down the values for better performance of ML algos
# Scaling --> two types --- 1. Standard Scaling --> Salary
# 2. MinMax Scaler

# It's a permanent step 

# 1. Standard Scaling (formula) --> [x-u(mean)/sigma]
# 2. MinMax Scaler (formula) --> [x-x(min)/x(max)-x(min)]

# Scaling is working only "Numerical Data"
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as ex
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```python
df1 = df[['Age', 'EstimatedSalary']]
```

```python
df1.describe()
```

```python
# Standard   ---> mean = 0, std = 1
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
df2_sc = sc.fit_transform(df1)
```

```python
df3 = pd.DataFrame(df2_sc, columns = ['Age', 'EstimatedSalary'])
```

```python
np.round(df3.describe(),2)
```
