```python
# Scaling -- to scale down the values for better performance of ML algos
   # Two types of Scaling --> 1. Standard Scaling, 2. MinMax Scaler
```

```python
# Scaling is a permanent process in ML
```

```python
# 1. Standard Scaling (formula) ==> [x-u(mean)/sigma]
# 2. MinMax Scaler (formula) ==> [x-x(min)/x(max)-x(min)]
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head(3)
```

```python
df1 = df[['Age', 'EstimatedSalary']]
df1.head(3)
```

```python
df1.describe()
```

```python
# we can see in df1.describe(), there are lots of up and downs in values, so we can say that here to level-up use scaling.
```

```python
# Using Scaling
```

```python
# Standard Scaling --> In it, mean = 0, std = 1.
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
df2_sc = sc.fit_transform(df1)
```

```python
df3 = pd.DataFrame(df2_sc, columns = ['Age', 'EstimatedSalary'])
```

```python
df3.head(3)
```

```python
df3.describe()
```

```python
np.round(df3.describe(), 2)
```

```python
# Scaling has completed for above dataset.
```
