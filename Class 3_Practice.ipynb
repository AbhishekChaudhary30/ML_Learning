```python
ds = pd.read_csv("loan_prediction.csv")
ds.head(3)
```

```python
ds.isnull().sum() # Checking missing values
```

```python
ds["Gender"].value_counts()
```

```python
ds["Married"].value_counts()
```

```python
ds["Dependents"].value_counts()
```

```python
ds["Self_Employed"].value_counts()
```

```python
ds["LoanAmount"].value_counts()
```

```python
ds["Loan_Amount_Term"].value_counts()
```

```python
ds["Credit_History"].value_counts()
```

```python
# Filling data in missing Values.
# Gender
ds.Gender.fillna(ds.Gender.mode()[0],inplace=True)

# Married
ds.Married.fillna(ds.Married.mode()[0],inplace=True)

# Dependents
ds.Dependents.fillna(ds.Dependents.mode()[0],inplace=True)

# Self_Employed
ds.Self_Employed.fillna(ds.Self_Employed.mode()[0],inplace=True)

# LoanAmount
ds.LoanAmount.fillna(ds.LoanAmount.median(),inplace=True)

# Loan_Amount_Term
ds.Loan_Amount_Term.fillna(ds.Loan_Amount_Term.median(), inplace=True)

# Credit_History
ds.Credit_History.fillna(ds.Credit_History.median(),inplace=True)
```

```python
ds.isnull().sum()
```

```python
ds.info()
```

```python
# Object Columns name -->
# "Gender", "Married", "Education", "Self_Employed", "Property_Area ", "Loan_Status".
```

```python
ds['Married'] = ds['Married'].map({"Yes":0, "No":1})
ds.head(10)
```

```python
ds['Education'].value_counts()
```

```python
ds['Education'] = ds['Education'].map({"Graduate":0, "Not Graduate":1})
ds.head(10)
```

```python
ds['Self_Employed'].value_counts()
```

```python
ds['Self_Employed'] = ds['Self_Employed'].map({"No":0, "Yes":1})
ds.head()
```

```python
ds['Property_Area'].value_counts()
```

```python
ds['Property_Area'] = ds['Property_Area'].map({"Semiurban":0, "Urban":1, "Rural":2})
ds.head()
```

```python
ds['Loan_Status'].value_counts()
```

```python
ds['Loan_Status'] =  ds['Loan_Status'].map({"Y":0, "N":1})
ds.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
```

```python
ds1 = pd.read_csv("loan_prediction.csv")
ds1.head()
```

```python
ds2 = ds1[["Gender", "Married", "Education", "Self_Employed", "Property_Area", "Loan_Status"]]
ds2.head()
```

```python
oe = OrdinalEncoder(categories = [["Male", "Female"], ["Yes", "No"], ["Graduate", "Not Graduate"], ["No", "Yes"], ["Semiurban", "Urban", "Rural"], ["Y", "N"]])
ds3 = oe.fit_transform(ds2)
ds3
```
